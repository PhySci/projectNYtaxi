```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_pickle('predictionResults.pcl')
```

```python
df.index.levels[0]
```

```python
f = open('myfile.txt', 'a')
for ts in df.index.levels[0]:
    for lag in df.index.levels[1][:-6]:
        for i in np.arange(6):
            try:
                s =  str(ts)+'_'+datetime.datetime.strftime(lag, "%Y-%m-%d") +'_'+ str(lag.hour)+'_'+str(i+1)+','+str(df.loc[ts,lag].y[i])+'\n'
                f.write(s)  # python will convert \n to os.linesep
            except Exception as ins:
                print ins
                break
                
f.close()            
```
