```python
testSet = pd.read_csv('data.csv')
testSet.set_index('data',inplace=True)
testSet.head()
```

```python
# id нужных регионов
regsDf = pd.read_csv('../crowdRegs.csv',names=['id','regId']);  

# временные ряды для этих регионов
mySet = pd.read_pickle('../loadData/crowdRegs3.pcl')
mySet.columns = regsDf.regId.values.astype('str')
mySet = mySet.loc['01-2016':'07-2016',:]
mySet.head()
```

```python
for colName in mySet.columns:
    mySeries = mySet.loc[:,colName]
    testSeries = testSet.loc[:,colName]
    diff = mySeries - testSeries
    diff = np.abs(diff)
    diff = np.sum(diff)
    print colName, diff
```
