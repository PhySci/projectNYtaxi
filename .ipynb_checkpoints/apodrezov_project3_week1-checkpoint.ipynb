# 2. Проведите фильтрацию: удалите поездки с

Удалим минуты и секунды во времени начала поездки:


```python
df['pickup_date_hour'] = [datetime.strptime(x[:-6], '%Y-%m-%d %H') for x in df['tpep_pickup_datetime']]
```

## 2.1 нулевой длительностью

Сначала посчитаем длительности поездок в секундах:

```python
df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])

df["trip_time_seconds"] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']) / np.timedelta64(1, 's')

df.shape
```

## 2.2 нулевым количеством пассажиров, нулевым расстоянием поездки по счётчику, координатами начала, не попадающими в прямоугольник Нью-Йорка

```python
pickup_long_start = -74.25559
pickup_long_end = -73.70001
pickup_lat_start = 40.49612
pickup_lat_end = 40.91553
```

```python
df_filtered = df[(df.trip_time_seconds != 0.) & (df.passenger_count != 0. ) & (df.trip_distance != 0.) &
                (df.pickup_longitude >= pickup_long_start) & 
                (df.pickup_longitude <= pickup_long_end) &
                (df.pickup_latitude >= pickup_lat_start) & 
                (df.pickup_latitude <= pickup_lat_end)]

df_filtered.shape
```

## 3. Нью-Йорк вписан в прямоугольник от -74.25559 до -73.70001 градусов долготы и от 40.49612 до 40.91553 широты. Разбейте этот прямоугольник на 2500 одинаковых прямоугольных областей — по 50 интервалов вдоль каждой оси.

```python
num_regions = 50

long_delta = (pickup_long_start - pickup_long_end) / num_regions
lat_delta = (pickup_lat_start - pickup_lat_end) / num_regions

ny_regions = {}

cnt = 1
for lng_index in range(1,51):
    lng_start = round(pickup_long_start - ((lng_index - 1) * long_delta), 7)
    lng_end = round(pickup_long_start - (lng_index * long_delta), 7)
    
    for lat_index in range(1,51):
        lt_start = round(pickup_lat_start - ((lat_index - 1) * lat_delta), 7)
        lt_end = round(pickup_lat_start - (lat_index * lat_delta), 7)      
        
        ny_regions[cnt] = [lng_start, lng_end, lt_start, lt_end]
        
        cnt += 1
```

Проверим нумерацию получившихся регионов и их координат из ny_regions и из файла regions.csv:

```python
import math

regions_df = pd.read_csv('regions.csv', sep = ';', index_col= 'region')

passed = 0
for index, row in regions_df.iterrows():

    if ((math.isclose(ny_regions[index][0], row['west'])) & 
        (math.isclose(ny_regions[index][1], row['east'])) &
        (math.isclose(ny_regions[index][2], row['south'])) & 
        (math.isclose(ny_regions[index][3], row['north']))):
            passed += 1
        
passed == 2500
```

Присвоим каждой записи в таблице номер региона:

```python
df_filtered["region"] = ((df_filtered['pickup_longitude']-pickup_long_start)/((pickup_long_end-pickup_long_start)/50)).astype(int)*50+1 +((df_filtered['pickup_latitude']-pickup_lat_start)/((pickup_lat_end-pickup_lat_start)/50)).astype(int)
```

## 4. Агрегируйте данные за май 2016 года по часам и областям (можно использовать функцию stats.binned_statistic_2d, с которой вы сталкивались в заданиях четвёртого курса). Не забудьте проверить, что идентификаторы, которые вы присваиваете ячейкам, соответствуют приложенному выше файлу.

Сначала агрегируем данные о количестве поездок по часам-регион:

```python
df_agg = pd.DataFrame(df_filtered.groupby(['region', 'pickup_date_hour'])["pickup_date_hour"].count().rename('trips'))
df_agg.head()
```

Теперь сделаем эту таблицу "плоской", т.е. избавимся от агрегации, но с сохранением всей информации. Вместо мульти-индекса (region, pickup_date_hour) добавим еще две колонки region, pickup_date_hour и продублируем для каждого значения trips значения в этих двух колонках:

```python
agg_list = []
for index, row in df_agg.iterrows():
    agg_list.append([index[0], index[1], row['trips']])
    
df_agg_flat = pd.DataFrame(agg_list, columns = ['region', 'pickup_date_hour', 'trips'])
df_agg_flat.head()
```

Теперь создадим пустой датафрейм с колонками - регион, дата начала поездки и количеством поездок - для каждой пары часы-регион за ВЕСЬ май 2016:

```python
%%time

lst=[]
for r in range(1,2501):        
    for d in range(1,32):
        for t in range(0,24):
            dat = datetime(2016, 5, d, t, 0, 0)
            trips = 0 #get_region_trips(r, dat)
            
            lst.append([r,dat,trips])
df_mask = pd.DataFrame(lst,columns=['region','pickup_date_hour', 'trips'])
df_mask.head()
```

Конкатенируем датафреймы df_agg_flat и df_mask. Таким образом, у нас получится список всех пар часы-регионы за май 2016, правда с дубликатами по колонкам region и  pickup_date_hour (для некоторых пар часы-регионы будет два значения trips - 0 и какое-то число N):

```python
data_for_agregate=pd.concat([df_mask, df_agg_flat], axis=0)  
data_for_agregate.shape
```

Чтобы избавиться от дубликатов и посчитать кол-во поездок для все пар часы-регион, опять сделаем агрегацию: 

```python
data_agregate=pd.DataFrame(data_for_agregate.groupby(['region','pickup_date_hour'])['trips'].sum())
```

## 6. Постройте график количества поездок жёлтого такси из ячейки, содержащей Empire State Building.

Найдем ячейку (регион), в который входит координаты Empire State Building:

```python
ESB = [40.748817, -73.985428]

ESB_region_id = 0

for key, value in ny_regions.items():
    pickup_long_start = value[0]
    pickup_long_end = value[1]
    pickup_lat_start = value[2]
    pickup_lat_end = value[3]
    
    if ((pickup_long_start <= ESB[1]) & (pickup_long_end >= ESB[1]) & (pickup_lat_start <= ESB[0]) & (pickup_lat_end >= ESB[0])):
        ESB_region_id = key
        break
        
print("Empire State Building region: ", ESB_region_id)
```

```python
plot_df = data_for_agregate[data_for_agregate.region == ESB_region_id][["pickup_date_hour","trips"]]
plot_df = plot_df.groupby(['pickup_date_hour'])['trips'].sum()

plt.figure(figsize=(20,7))
plot_df.plot(kind='line')
plt.ylabel('Trips count')
plt.xlabel('Date')
plt.title('Trips from Empire State Building block')
```
