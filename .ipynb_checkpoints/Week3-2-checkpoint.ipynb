```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn import linear_model
%matplotlib inline
```

```python
engine = create_engine('postgresql://user:user@localhost/ML_project2');
```

```python
startDate = datetime(2014,1,1,0,0,0);
endDate = datetime(2016,5,30,23,59,59);
```

```python
if True:
    # делаем запрос к базе и дёргаем статистику по желаемым датам и региону
    query = "SELECT date_trunc('hour',trip_pickup_datetime) AS date, count(trip_pickup_datetime) ";
    query = query + "FROM trip WHERE regions= %(region)s AND ";
    query = query + "trip_pickup_datetime BETWEEN %(startDate)s AND %(endDate)s "; 
    query = query + "GROUP BY date_trunc('hour',trip_pickup_datetime) ";
    query = query +"ORDER BY date_trunc('hour',trip_pickup_datetime) "
    res = pd.read_sql_query(query,engine, params = {"region":1734,"startDate":startDate,"endDate":endDate}, index_col = 'date')
    res.to_pickle('region-1734-2014-16')
```

```python
res = pd.read_pickle('region-1734-2014-16')
## fill empty hours 
date_index = pd.date_range(startDate, endDate, freq='H')
res = res.reindex(date_index,fill_value =0)
res.plot()

# add linear feature
res = res.assign(hours = (res.index-res.index[0])/np.timedelta64(1, 'h'))
```

```python
# add week periodicity
K = 6

for ind in range(1,K+1):
    res['weekCos'+str(ind)]= np.cos(np.pi*res.hours*ind/168);
    res['weekSin'+str(ind)]= np.sin(np.pi*res.hours*ind/168);

res[0:168].plot('hours',['weekCos1','weekCos2','weekCos3','weekCos4']);
```

```python
# add annual periodicity
K = 3

for ind in range(1,K+1):
    res['yearCos'+str(ind)]= np.cos(2*np.pi*res.hours*ind/8766);
    res['yearSin'+str(ind)]= np.sin(2*np.pi*res.hours*ind/8766);

res[0:8766].plot('hours',['yearCos1','yearCos2']);
```

## Lasso regression

```python
def lassoRegression(df):
    X = df.drop('count',axis = 1);
    y = df.loc[:,'count'];
    lassoReg = linear_model.Lasso(alpha = 0.1, max_iter = 1e5,fit_intercept=True);
    lassoReg.fit(X,y)
    y_pr = pd.Series(data = lassoReg.predict(X), index = df.index)
    R = lassoReg.score(X,y);
    print 'R factor is ', R
    
    plt.figure(figsize = [10,8])
    plt.plot(y)
    plt.plot(y-y_pr)
    plt.legend(['original data','residuals'])
    return [y_pr, y-y_pr] 
```

```python
[s,r]=lassoRegression(res)
```

Добавим дни недели, чтобы улучшить фит недельной сезонности

```python
#res.merge() 
DoW = pd.get_dummies(res.index.dayofweek,drop_first= True,prefix = 'dayofWeek');
DoW.index = date_index;
res = res.merge(DoW,left_index=True,right_index=True)
lassoRegression(res)
```

Попробуем подобрать сезонные коэффициенты для среднесуточных значений количества поездок. 
А потом будем использовать их для модели ARIMA.

```python
res2 = pd.read_pickle('region-1734-2014-16');
res2 = res2.groupby(pd.DatetimeIndex(res2.index).round('1D') ).agg('mean')
```

```python
# add linear feature
res2['days'] = (res2.index-res2.index[0])/np.timedelta64(1, 'D')
```

```python
# add week periodicity
K = 5

for ind in range(1,K+1):
    res2['weekCos'+str(ind)]= np.cos(2*np.pi*res2.days*ind/7);
    res2['weekSin'+str(ind)]= np.sin(2*np.pi*res2.days*ind/7);

# add annual periodicity
K = 3

for ind in range(1,K+1):
    res2['yearCos'+str(ind)]= np.cos(2*np.pi*res2.days*ind/8766);
    res2['yearSin'+str(ind)]= np.sin(2*np.pi*res2.days*ind/8766);
```

```python
[s1,r1] = lassoRegression(res2)
```

Добавим дни недели

```python
DoW = pd.get_dummies(res2.index.dayofweek,drop_first= True,prefix = 'dayofWeek');
DoW.index = res2.index;
res2 = res2.merge(DoW,left_index=True,right_index=True)
[s2,r2] = lassoRegression(res2)
```

```python
# добавим месяцы для учёта годовой сезоности
mth = pd.get_dummies(res2.index.month,drop_first= True,prefix = 'month');
mth.index = res2.index;
res2 = res2.merge(mth,left_index=True,right_index=True);
[s3,r3] = lassoRegression(res2)
```

```python
plt.figure(figsize=[16,8]);
plt.plot(r1)
plt.plot(r3)
plt.legend(['series 1','series 3'])
plt.xlim(['2014-1','2016-6'])
```
